Mengunduh File Haar Cascade dari Url

```python
import urllib.request

# url untuk file haarcascade dari OpenCV
url = 'https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml'

# menyimpan file haar cascade di google colab
urllib.request.urlretrieve(url, '/content/haarcascade_frontalface_default.xml')
```

Membuka Gambar dan dijadikan skala pada plot


Gambar yang disimpan di Google Drive akan ditampilkan di Google Colab menggunakan matplotlib. Gambar akan tampil dengan warna yang benar karena telah dikonversi ke format RGB dari BGR yang digunakan oleh OpenCV.

```python
import cv2
from matplotlib import pyplot as plt
from google.colab import drive

# mounting ke google drive
drive.mount('/content/drive')

# membaca gambar yang disimpan pada google drive
img = cv2.imread('/content/drive/MyDrive/DataScience/Gambar/timnas.jpeg')

# mengkonversi gambar ke grayscale dan rgb
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# menampilkan gambar menggunakan matplotlib
plt.subplot(1,1,1)
plt.imshow(img_rgb)
plt.show()
```

Recognation

Kode ini digunakan untuk deteksi objek dalam gambar menggunakan metode Haar Cascade Classifier yang telah dilatih sebelumnya.

```python
# membuat Clasifier Haar Cascade
stop_data = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')

# mendeteksi objek pada gambar
found = stop_data.detectMultiScale(img_gray,
                                   minSize =(20, 20))

#menghitung jumlah objek yang ditemukan
amount_found = len(found)


# menandai objek yang ditemukan dengan kotak hijau
if amount_found != 0:
  for (x, y, width, height) in found:
    cv2.rectangle(img_rgb, (x, y),
                      (x + height, y + width),
                      (0, 255, 0), 5)

plt.subplot(1, 1, 1)
plt.imshow(img_rgb)
plt.show()
```
